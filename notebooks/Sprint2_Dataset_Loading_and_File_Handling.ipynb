# Module 8 – Custom Exception Handling

## Purpose

Exception handling ensures that the notebook can gracefully handle unexpected situations such as:

- Missing files
- Invalid file formats
- Corrupted datasets
- Invalid worksheet names
- Incorrect file paths

Instead of displaying generic Python exceptions, meaningful custom exceptions are created to improve debugging and readability.

````python
class DatasetLoadingException(Exception):
    """
    Base exception for dataset loading errors.
    """
    pass


class InvalidFilePathException(DatasetLoadingException):
    """
    Raised when the file path is invalid.
    """
    pass


class UnsupportedFileFormatException(DatasetLoadingException):
    """
    Raised when an unsupported file format is used.
    """
    pass


class InvalidWorksheetException(DatasetLoadingException):
    """
    Raised when an Excel worksheet does not exist.
    """
    pass


class CorruptedFileException(DatasetLoadingException):
    """
    Raised when a file is corrupted.
    """
    pass
````

````python
try:
    raise InvalidFilePathException(
        "Dataset path does not exist."
    )

except InvalidFilePathException as error:
    logger.error(error)
    print(error)
````

````python
try:
    raise UnsupportedFileFormatException(
        "XML files are not supported."
    )

except UnsupportedFileFormatException as error:
    logger.error(error)
    print(error)
````

# Module 5 – Dataset Writer

## Purpose

After loading and processing datasets, the next step in any data pipeline is to save the processed data.

This module demonstrates how to write datasets into multiple formats while maintaining reusable, modular, and readable code.

Supported Output Formats:

- CSV
- JSON
- Excel
- TXT
- Pickle

All writing operations are logged, validated, and protected with exception handling.

````python
def save_csv(dataframe: pd.DataFrame,
             output_path: Path) -> None:
    """
    Save DataFrame as CSV.
    """

    try:

        logger.info(f"Saving CSV : {output_path}")

        dataframe.to_csv(
            output_path,
            index=False
        )

        logger.info("CSV saved successfully.")

    except Exception as error:

        logger.error(error)

        raise
````

````python
def save_json(dataframe: pd.DataFrame,
              output_path: Path) -> None:
    """
    Save DataFrame as JSON.
    """

    try:

        logger.info(f"Saving JSON : {output_path}")

        dataframe.to_json(
            output_path,
            orient="records",
            indent=4
        )

        logger.info("JSON saved successfully.")

    except Exception as error:

        logger.error(error)

        raise
````

````python
def save_excel(dataframe: pd.DataFrame,
               output_path: Path) -> None:
    """
    Save DataFrame as Excel.
    """

    try:

        logger.info(f"Saving Excel : {output_path}")

        dataframe.to_excel(
            output_path,
            index=False
        )

        logger.info("Excel saved successfully.")

    except Exception as error:

        logger.error(error)

        raise
````

````python
def save_txt(dataframe: pd.DataFrame,
             output_path: Path) -> None:
    """
    Save DataFrame as TXT.
    """

    try:

        logger.info(f"Saving TXT : {output_path}")

        with open(
            output_path,
            "w",
            encoding="utf-8"
        ) as file:

            file.write(
                dataframe.to_string(
                    index=False
                )
            )

        logger.info("TXT saved successfully.")

    except Exception as error:

        logger.error(error)

        raise
````

````python
def save_pickle(dataframe: pd.DataFrame,
                output_path: Path) -> None:
    """
    Save DataFrame as Pickle.
    """

    try:

        logger.info(f"Saving Pickle : {output_path}")

        with open(
            output_path,
            "wb"
        ) as file:

            pickle.dump(
                dataframe,
                file
            )

        logger.info("Pickle saved successfully.")

    except Exception as error:

        logger.error(error)

        raise
````

````python
save_csv(
    employees_df,
    OUTPUT_DIR / "employees_output.csv"
)

save_json(
    employees_df,
    OUTPUT_DIR / "employees_output.json"
)

save_excel(
    employees_df,
    OUTPUT_DIR / "employees_output.xlsx"
)

save_txt(
    employees_df,
    OUTPUT_DIR / "employees_output.txt"
)

save_pickle(
    employees_df,
    OUTPUT_DIR / "employees_output.pkl"
)

print("All datasets saved successfully.")
````

# Module 6 – File Path Management

## Purpose

Managing file paths correctly is essential for building reusable and portable data pipelines.

Instead of hardcoding file locations throughout the notebook, reusable helper functions are created to:

- Validate file paths
- Create directories automatically
- Generate output filenames
- Work with both relative and absolute paths

Using `pathlib` improves readability and ensures compatibility across different operating systems.

````python
def validate_file_path(file_path: Path) -> bool:
    """
    Validate whether the file exists.

    Parameters
    ----------
    file_path : Path
        Path of the input file.

    Returns
    -------
    bool
        True if file exists.

    Raises
    ------
    InvalidFilePathException
        If file does not exist.
    """

    if not file_path.exists():

        logger.error(f"File not found : {file_path}")

        raise InvalidFilePathException(
            f"{file_path} does not exist."
        )

    logger.info(f"Validated file : {file_path}")

    return True
````

````python
validate_file_path(csv_file)
````

````python
def create_directory(directory: Path) -> None:
    """
    Create directory if it does not exist.
    """

    directory.mkdir(
        parents=True,
        exist_ok=True
    )

    logger.info(f"Directory verified : {directory}")
````

````python
create_directory(OUTPUT_DIR)

print("Directory Ready")
````

````python
from datetime import datetime

def generate_output_filename(
    prefix: str,
    extension: str
) -> Path:
    """
    Generate output filename with timestamp.
    """

    timestamp = datetime.now().strftime(
        "%Y%m%d_%H%M%S"
    )

    filename = (
        f"{prefix}_{timestamp}.{extension}"
    )

    return OUTPUT_DIR / filename
````

````python
def log_start(operation: str):

    logger.info("=" * 60)

    logger.info(f"STARTED : {operation}")

    logger.info("=" * 60)


def log_success(operation: str):

    logger.info(f"SUCCESS : {operation}")


def log_warning(message: str):

    logger.warning(message)


def log_error(error):

    logger.error(error)
````

````python
def save_all_formats(dataframe: pd.DataFrame):
    """
    Save a dataset in all supported formats.
    """

    save_operations = [
        ("CSV", save_csv, OUTPUT_DIR / "employees_output.csv"),
        ("JSON", save_json, OUTPUT_DIR / "employees_output.json"),
        ("Excel", save_excel, OUTPUT_DIR / "employees_output.xlsx"),
        ("TXT", save_txt, OUTPUT_DIR / "employees_output.txt"),
        ("Pickle", save_pickle, OUTPUT_DIR / "employees_output.pkl"),
    ]

    for name, writer_function, output_path in save_operations:
        log_start(f"Saving {name}")

        writer_function(dataframe, output_path)

        log_success(f"Saving {name}")

    print("All datasets saved successfully.")
````

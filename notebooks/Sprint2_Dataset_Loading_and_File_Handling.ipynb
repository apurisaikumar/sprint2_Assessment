# Module 7 – Logging Configuration

## Purpose

Logging is an essential software engineering practice that helps monitor application execution, debug issues, and maintain an execution history.

In this notebook, logging is configured to:

- Track file loading operations
- Track file writing operations
- Record warning and error messages
- Capture exception details
- Store logs in `logs/application.log`

This centralized logging mechanism improves traceability and simplifies debugging.

````python
import logging

LOG_FILE = LOG_DIR / "application.log"

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s | %(levelname)s | %(message)s",
    handlers=[
        logging.FileHandler(LOG_FILE),
        logging.StreamHandler()
    ],
    force=True
)

logger = logging.getLogger(__name__)

logger.info("Logging configured successfully.")
````

````python
logger.info("Dataset loading started.")

logger.warning("This is a sample warning message.")

logger.error("This is a sample error message.")
````

# Module 8 – Custom Exception Handling

## Purpose

Exception handling ensures that the notebook can gracefully handle unexpected situations such as:

- Missing files
- Invalid file formats
- Corrupted datasets
- Invalid worksheet names
- Incorrect file paths

Instead of displaying generic Python exceptions, meaningful custom exceptions are created to improve debugging and readability.

````python
class DatasetLoadingException(Exception):
    """
    Base exception for dataset loading errors.
    """
    pass


class InvalidFilePathException(DatasetLoadingException):
    """
    Raised when the file path is invalid.
    """
    pass


class UnsupportedFileFormatException(DatasetLoadingException):
    """
    Raised when an unsupported file format is used.
    """
    pass


class InvalidWorksheetException(DatasetLoadingException):
    """
    Raised when an Excel worksheet does not exist.
    """
    pass


class CorruptedFileException(DatasetLoadingException):
    """
    Raised when a file is corrupted.
    """
    pass
````

````python
try:
    raise InvalidFilePathException(
        "Dataset path does not exist."
    )

except InvalidFilePathException as error:
    logger.error(error)
    print(error)
````

````python
try:
    raise UnsupportedFileFormatException(
        "XML files are not supported."
    )

except UnsupportedFileFormatException as error:
    logger.error(error)
    print(error)
````
